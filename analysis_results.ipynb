```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
width = height = 240
discount = 0.9
max_memory = 100000
learning_rate = 0.001
batch_size = 1000

def_results_file = f'model_w={width}_h={height}_d={discount}_m={max_memory}_lr={learning_rate}_b={batch_size}.csv'
```

```python
def_data = (
    pd.read_csv(os.path.join('results', def_results_file))
    .transpose()
    .reset_index()
    .reset_index()
    .astype(float)
)
def_data.columns = ['iter', 'score', 'cumulative_score']
```

```python
def load_data(width, height, discount, max_memory, learning_rate, batch_size):
    results_file = f'model_w={width}_h={height}_d={discount}_m={max_memory}_lr={learning_rate}_b={batch_size}.csv'
    data = (
        pd.read_csv(os.path.join('results', results_file))
        .transpose()
        .reset_index()
        .reset_index()
        .astype(float)
    )
    data.columns = ['iter', 'score', 'cumulative_score']
    return data
```

# Batch size

```python
sns.lineplot(def_data, x='iter', y='score', color='blue', linewidth=0.5)
sns.lineplot(def_data, x='iter', y='cumulative_score', color='blue', linewidth=3, label='Batch size = 1000')

_ = load_data(width, height, discount, max_memory, learning_rate, 2000)
sns.lineplot(_, x='iter', y='score', color='orange', linewidth=0.5)
sns.lineplot(_, x='iter', y='cumulative_score', color='orange', linewidth=3, label='Batch size = 2000')

_ = load_data(width, height, discount, max_memory, learning_rate, 500)
sns.lineplot(_, x='iter', y='score', color='green', linewidth=0.5)
sns.lineplot(_, x='iter', y='cumulative_score', color='green', linewidth=3, label='Batch size = 500')

plt.title('Effect of Batch Size')
plt.xlabel('Number of Games')
plt.ylabel('Score')

plt.savefig('images/batch_size.png')

plt.show()
```
